```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
print(cancao)
cancao = cancao.casefold() #transforma a primeira letra em minusculo
qtd_roda = cancao.count("roda") #conta quantas vezes tem a palavra roda
print(qtd_roda)
cancao = cancao.split(',') #a cada virgula, separa em uma string diferente
print(cancao[0])
print(cancao[1])
print(cancao[2])
print(cancao[3])
```
