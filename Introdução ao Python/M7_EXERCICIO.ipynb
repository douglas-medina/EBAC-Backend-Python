```python
import os

os.rename('./dados/dow_jones_index.data', './dados/dow_jones_index.csv')
```

```python
!pip install pandas==1.1.5
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./dados/dow_jones_index.csv')
```

```python
df.head(n=10)
```

```python
df.columns.to_list()
```

```python
linhas, colunas = df.shape
print(f'Número de linhas: {linhas}')
print(f'Número de colunas: {colunas}')
```

```python
df_mcd = df[df['stock'] == 'MCD']
```

```python
df_mcd.head()
```

```python
df_mcd = df_mcd[['date', 'open', 'high', 'low', 'close']]
```

```python
df_mcd.head(n=10)
```

```python
df_mcd.dtypes
```

```python
for col in ['open', 'high', 'low', 'close']:
  df_mcd[col] = df_mcd[col].apply(lambda value: float(value.split(sep='$')[-1]))
```

```python
df_mcd.head()
```

```python
df_mcd.dtypes
```

```python
df_coke = df[df['stock'] == 'KO']
```

```python
df_coke = df_coke[['date', 'open', 'high', 'low', 'close']]
```

```python
for col in ['open', 'high', 'low', 'close']:
  df_coke[col] = df_coke[col].apply(lambda value: float(value.split(sep='$')[-1]))
```

```python
df_coke.head()
```

# SEABORN

```python
!pip install seaborn==0.11.1
```

```python
import seaborn as sns
```

valores de abertura das ações

```python
plot = sns.lineplot(x="date", y="open", data=df_mcd)
_ = plot.set_xticklabels(labels=df_mcd['date'], rotation=90)
```

valores de fechamento das ações

```python
plot = sns.lineplot(x="date", y="close", data=df_mcd)
_ = plot.set_xticklabels(labels=df_mcd['date'], rotation=90)
```

4 valores no mesmo gráfico

```python
plot = sns.lineplot(x="date", y="value", hue='variable', data=pd.melt(df_mcd, ['date']))
_ = plot.set_xticklabels(labels=df_mcd['date'], rotation=90)
```

salvando o gráfico como png

```python
plot.figure.savefig("./mcd.png")
```

```python
plot_coke = sns.lineplot(x="date", y="open", data=df_coke)
_ = plot.set_xticklabels(labels=df_coke['date'], rotation=90)
```

```python
plot_coke = sns.lineplot(x="date", y="close", data=df_coke)
_ = plot.set_xticklabels(labels=df_coke['date'], rotation=90)
```

```python
plot_coke = sns.lineplot(x="date", y="value", hue='variable', data=pd.melt(df_coke, ['date']))
_ = plot.set_xticklabels(labels=df_coke['date'], rotation=90)
```

```python
plot_coke.figure.savefig("./coke_chart.png")
```

# Insight

Mesmo sabendo da variação que pode haver, o mcdonalds ainda se mantém sempre em uma média de valores. Enquanto a Coca Cola dificilmente mantém uma média num dia, menos ainda num grande período de tempo.

Num periodo de 6 meses o mcdonalds conseguiu progressivamente auemntar os valores das suas ações, tendo um aumento significativo no ultimo período. Enquanto a coca cola com suas altas e baixas, terminou o período com praticamente o mesmo valor inicial por ação.
