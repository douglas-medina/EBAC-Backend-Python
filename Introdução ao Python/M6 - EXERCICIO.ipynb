EXERCICIO ARQUIVO DE TEXTO

```python
class ArquivoTexto(object):
  def __init__(self, arquivo: str):
    self.arquivo = arquivo
    self.conteudo = self._extrair_conteudo()

  def _extrair_conteudo(self):
      conteudo = None
      with open(file=self.arquivo, mode='r', encoding='utf8') as arquivo:
          conteudo = arquivo.readlines()
      return conteudo

  def linha(self, numero_linha: int):
    if numero_linha < 1 or numero_linha > len(self.conteudo):
      raise ValueError(f"Número de linha inválido: {numero_linha}")

    return self.conteudo[numero_linha - 1].rstrip("\n")
```

```python
arquivo_texto = ArquivoTexto(arquivo='musica.txt')
print(arquivo_texto.arquivo)

numero_linha = 1
print(arquivo_texto.linha(numero_linha=numero_linha))
# Roda Viva

numero_linha = 9
print(arquivo_texto.linha(numero_linha=numero_linha))
# Mas eis que chega a roda viva
```

EXERCICIO ARQUIVO CSV

```python
class ArquivoCSV(ArquivoTexto):
  def __init__(self, arquivo: str):
    super().__init__(arquivo=arquivo)
    self.colunas = self._extrair_nome_colunas()

  def _extrair_nome_colunas(self):
      return self.conteudo[0].strip().split(sep=',')

  def extrair_coluna_da_linha(self, numero_coluna: int):
    super().linha(linha=numero_linha)
    coluna = list()
    for linha in self.conteudo:
      conteudo_linha = linha
      coluna.append(conteudo_linha[numero_coluna])
      coluna.pop(0)
    return coluna
```

```python
arquivo_csv = ArquivoCSV(arquivo='carros.csv')
print(arquivo_csv)

numero_linha = 1
print(arquivo_csv.linha(numero_linha=numero_linha))
# id,valor_venda,valor_manutencao,portas,pessoas,porta_malas

print(arquivo_csv.colunas)
# [
# 'id',
# 'valor_venda',
# 'valor_manutencao',
# 'portas',
# 'pessoas',
# 'porta_malas'
# ]

numero_linha = 10
print(arquivo_csv.linha(numero_linha=numero_linha))
# 9,low,med,2,2,small
```
