```python
emprestimos = []
with open(file='./credito.csv', mode='r', encoding='utf8') as fp:
    fp.readline()  # cabeçalho
    linha = fp.readline()
    while linha:
        linha_elementos = linha.strip().split(sep=',')
        linha_emprestimo = {
            'id_vendedor': linha_elementos[0],
            'valor_emprestimos': linha_elementos[1],
            'quantidade_emprestimos': linha_elementos[2],
            'data': linha_elementos[3]
        }
        emprestimos.append(linha_emprestimo)
        linha = fp.readline()  # ler próxima linha
```

```python
for emprestimo in emprestimos:
 print(emprestimo)
```

```python
valor_emprestimos_lista = list(map(lambda valor: float(valor['valor_emprestimos']), emprestimos))

print(valor_emprestimos_lista)
```

```python
valor_emprestimos_lista_filtrada = list(filter(lambda valor: valor >= 0, valor_emprestimos_lista))
print(valor_emprestimos_lista_filtrada)
```

```python
from functools import reduce

soma_valores_emprestimos = reduce(lambda valor1, valor2: valor1 + valor2, valor_emprestimos_lista_filtrada)
print(soma_valores_emprestimos)
```

```python
from functools import reduce

quantidade = len(valor_emprestimos_lista)
print(quantidade)

media = reduce(lambda x, y: x + y, valor_emprestimos_lista_filtrada) / len(valor_emprestimos_lista_filtrada)
print(media)
```

```python
from functools import reduce
import math

media = reduce(lambda x, y: x + y, valor_emprestimos_lista_filtrada) / len(valor_emprestimos_lista_filtrada)
soma_diferencas_quadradas = reduce(lambda x, y: x + (y - media) ** 2, valor_emprestimos_lista_filtrada)

variancia = soma_diferencas_quadradas / len(valor_emprestimos_lista_filtrada)
desvio_padrao_valor_emprestimos = math.sqrt(variancia)

print(desvio_padrao_valor_emprestimos)
```
